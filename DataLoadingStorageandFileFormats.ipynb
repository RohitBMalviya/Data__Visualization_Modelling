```python
unified_data = pd.merge(dfcsv, dfexcel, how='outer')
unified_data.fillna(0)
```

```python
unified_data['ORDERDATE'] = pd.to_datetime(unified_data['ORDERDATE'])
unified_data['year'] = unified_data['ORDERDATE'].dt.year
unified_data['month'] = unified_data['ORDERDATE'].dt.month
unified_data['day'] = unified_data['ORDERDATE'].dt.day
unified_data
```

```python
merged_data = pd.merge(dfcsv, dfexcel, how='inner')
```

```python
unified_data.describe()
```

```python
sales_by_country = unified_data.groupby('COUNTRY')['SALES'].sum()
```

```python
pd.DataFrame(sales_by_country)
```

```python
average_order_value = unified_data['SALES'].mean()
average_order_value
```

```python
import matplotlib.pyplot as plt
sales_by_country.plot(kind='bar', title='Sales by Country')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.show()
```
