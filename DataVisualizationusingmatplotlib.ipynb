```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('city_day.csv')
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['AQI'])
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('Overall AQI Trend Over Time')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['PM2.5'], label='PM2.5')
plt.plot(df['Date'], df['PM10'], label='PM10')
plt.plot(df['Date'], df['CO'], label='CO')
plt.xlabel('Date')
plt.ylabel('Pollutant Levels')
plt.title('Pollutant Trends Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Date'], df['AQI'])
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('AQI Values Across Dates')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df.boxplot(column='AQI', by='AQI_Bucket', patch_artist=True, vert=False, showfliers=False)
plt.xlabel('Pollutant Category')
plt.ylabel('AQI')
plt.title('AQI Distribution Across Pollutant Categories')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['PM2.5'], df['AQI'])
plt.xlabel('PM2.5 Levels')
plt.ylabel('AQI')
plt.title('Scatter Plot: PM2.5 Levels vs AQI')
plt.show()
```
